### Overview
이 프로젝트는 NASA 소행성 데이터를 기반으로 지구 근접 천체`neo`와 잠재적 지구위협 소행성`pha`을 분류하는 것을 목표로 한다.
원본 데이터에는 소행성의 물리적 특성, 궤도 요소, 식별자, 파생 변수 등이 포함되어 있다. 모델 학습에 앞서 target leakage 가능성이 있거나, 식별자, 상수 데이터 및 결측치가 많거나 다른 변수로부터 유도되는 feature를 제거하였다.

# 1. Data Preprocessing
## 1.1. 데이터 개요
원본 데이터는 71,351개의 소행성 관측치와 32개의 feature로 구성되어 있다. 전처리 후에는 70,578개의 행과 9개의 feature가 남았다.

| 항목              |            값 |
| --------------- | -----------: |
| 원본 행 수          |       71,351 |
| 원본 feature 수    |           32 |
| 전처리 후 행 수       |       70,578 |
| 전처리 후 feature 수 |            9 |
| Target variable | `neo`, `pha` |



이 프로젝트의 분류 흐름은 2단계로 구성된다. 먼저 `neo`를 예측하여 지구 근접 천체 여부를 판단한다. `neo = N`으로 분류된 경우 추가적인 위험 천체 판정을 수행하지 않고, `neo = Y`로 분류된 천체에 대해서만 `pha`를 추가로 예측한다.

## 1.2. Feature 요약

각 feature에 대해 데이터 타입, 결측치 개수, 결측률, 고유값 개수, 상수 여부를 확인하였다. 이를 통해 제거할 feature를 선정하였다.

Feature 요약 결과 `G`, `PC`, `diameter`, `extent`, `albedo`, `rot_per`, `BV`는 결측률이 90% 이상이었다. 해당 feature들은 관측값이 거의 존재하지 않아 모델 학습에 사용하기 어렵다고 판단하여 열 단위로 삭제하였다.

전처리 후 남은 feature는 다음과 같다.

| Feature    | 설명               |
| ---------- | ---------------- |
| `neo`      | 지구 근접 천체 여부      |
| `pha`      | 잠재적 지구위협 소행성 여부  |
| `sats`     | 위성 수             |
| `e`        | 이심률              |
| `a`        | 궤도 장반경           |
| `i`        | 궤도 경사            |
| `om`       | 승교점 경도           |
| `w`        | 근일점 인수           |
| `moid_jup` | 목성과의 최소 궤도 교차 거리 |

결측률 외에도 식별자, 상수 feature, 파생 변수, target leakage 가능성이 있는 feature를 제거하였다. 제거한 Feature는 다음과 같다.

| 제거 feature                   | 제거 이유                                            |
| ---------------------------- | ------------------------------------------------ |
| `spkid`, `full_name`, `name` | 소행성 고유 식별자 또는 이름                                 |
| `orbit_id`                   | NASA/JPL 내부 궤도해 식별자                              |
| `equinox`                    | 모든 값이 동일한 상수 feature                             |
| `tp`, `tp_cal`               | 근일점 통과 시각과 그 달력형 변환값                             |
| `moid`, `moid_ld`            | `pha` 정의에 직접 관련된 변수이며, `moid_ld`는 `moid`의 단위 변환값 |
| `H`                          | `pha` 정의에 직접 관련된 변수                              |
| `class`                      | 궤도 특성을 바탕으로 부여된 분류 label                         |
| `q`, `ad`                    | `a`, `e`로부터 유도 가능한 변수                            |
| `n`, `per`, `per_y`          | 공전 주기 및 공전 각속도 관련 중복 변수                          |

일부 변수는 다른 궤도 요소로부터 계산할 수 있다. 예를 들어 근일점 거리 `q`와 원일점 거리 `ad`는 다음과 같이 계산된다.

$$
q = a(1 - e)
$$

$$
ad = a(1 + e)
$$

따라서 `q`와 `ad`는 중복 정보를 줄이기 위해 제거하였다. `per`, `per_y`, `n` 역시 공전 주기와 공전 각속도를 나타내는 중복 변수이므로 제거하였다.

# 2. Exploratory Data Analysis(EDA)

전처리 후 데이터는 70,578개의 행과 9개의 feature로 구성되어 있다. 결측치가 많은 열과 중복·파생·leakage 가능성이 있는 feature를 제거한 뒤, 남은 데이터에서 `neo`, `pha`의 분포와 각 feature의 특성을 확인하였다.

## 2.1. Target 분포

`neo`는 지구 근접 천체 여부를 나타내며, 전처리 후 데이터에서는 `Y`가 41,644개, `N`이 28,934개로 확인되었다.

| `neo` |     개수 |     비율 |
| ----- | -----: | -----: |
| `N`   | 28,934 | 41.00% |
| `Y`   | 41,644 | 59.00% |

`neo`는 `Y`가 더 많지만, 두 클래스의 차이가 극단적이지는 않다. 반면 `pha`는 매우 불균형한 분포를 보인다.

| `pha` |     개수 |     비율 |
| ----- | -----: | -----: |
| `N`   | 68,032 | 96.39% |
| `Y`   |  2,546 |  3.61% |

`pha = Y`는 전체 데이터의 3.61%에 불과하다. 따라서 `pha` 분류 모델을 학습할 때는 클래스 불균형을 반드시 고려해야 한다. Accuracy만 평가 지표로 사용할 경우 대부분을 `pha = N`으로 예측해도 높은 성능처럼 보일 수 있으므로, Precision, Recall, F1-score, ROC-AUC, PR-AUC 등을 함께 확인할 필요가 있다.


## 2.2. `neo`와 `pha`의 관계

`neo`와 `pha`의 교차 분포는 다음과 같다.

| `neo` | `pha = N` | `pha = Y` |
| ----- | --------: | --------: |
| `N`   |    28,934 |         0 |
| `Y`   |    39,098 |     2,546 |

`neo = N`인 천체 중 `pha = Y`인 경우는 존재하지 않았다. 즉, 잠재적 지구위협 소행성(`pha = Y`)은 모두 지구 근접 천체(`neo = Y`)에 포함되어 있다.

따라서 본 프로젝트에서는 먼저 `neo` 여부를 분류하고, `neo = Y`로 분류된 천체에 대해서만 추가로 `pha` 여부를 판단하는 2단계 분류 구조를 사용한다.

다만 `neo = Y`인 천체 내부에서도 `pha = Y`는 2,546개로, `neo = Y` 데이터의 6.11%에 불과하다.

| `pha` in `neo = Y` |     개수 |     비율 |
| ------------------ | -----: | -----: |
| `N`                | 39,098 | 93.89% |
| `Y`                |  2,546 |  6.11% |

따라서 2단계 모델에서도 `pha` 분류는 여전히 클래스 불균형이 큰 문제로 남는다.

## 2.3 Feature별 특징

전처리 후 남은 입력 feature는 `sats`, `e`, `a`, `i`, `om`, `w`, `moid_jup`이다. `neo`와 `pha`는 target으로 사용된다.

| Feature    | 주요 특징                                                                      |
| ---------- | -------------------------------------------------------------------------- |
| `sats`     | 대부분의 천체가 위성을 갖지 않는다. `sats = 0`이 70,455개로 전체의 99.83%를 차지한다.                |
| `e`        | 이심률은 평균 0.379, 중앙값 0.368이다. `pha = Y`의 중앙값은 0.566으로 `pha = N`의 0.363보다 높다. |
| `a`        | 궤도 장반경은 중앙값 2.026 AU이다. 최대값이 350.3 AU로 매우 큰 이상치가 존재한다.                     |
| `i`        | 궤도 경사는 중앙값 9.25도이다. 최대값은 165.6도로, 일부 고경사 궤도 천체가 포함되어 있다.                   |
| `om`       | 승교점 경도는 0도에서 360도 범위의 각도형 변수이다. 특정 target에 따라 뚜렷한 평균 차이는 크지 않다.            |
| `w`        | 근일점 인수 역시 0도에서 360도 범위의 각도형 변수이다. `om`과 마찬가지로 궤도의 방향 정보를 나타낸다.             |
| `moid_jup` | 목성과의 최소 궤도 교차 거리이다. 평균은 2.605 AU, 중앙값은 2.56 AU이다.                          |

`e`는 target과의 차이가 비교적 뚜렷하게 나타났다. `neo = Y`의 이심률 중앙값은 0.448로 `neo = N`의 0.304보다 높았고, `pha = Y`의 이심률 중앙값도 0.566으로 높게 나타났다. 이는 위험 천체 분류에서 궤도 형태가 중요한 정보가 될 수 있음을 보여준다.

`a`와 `i`는 일부 극단값이 존재하므로, 딥러닝 모델 학습 전 feature scaling을 적용하는 것이 필요하다. 특히 `a`는 최대값이 350.3 AU로 중앙값 2.026 AU와 차이가 크다.

`om`과 `w`는 각도형 변수이므로 일반적인 연속형 숫자와는 다르게 해석해야 한다. 두 변수 모두 0도와 360도가 연결되는 순환 구조를 가지므로, 필요하다면 이후 모델링 과정에서 sin/cos 변환을 고려할 수 있다.


## 2.4 EDA 요약

전처리 후 데이터에서는 `neo`보다 `pha`의 클래스 불균형이 훨씬 심하게 나타났다. `pha = Y`는 전체의 3.61%에 불과하며, `neo = Y`로 제한해도 6.11%만 `pha = Y`에 해당한다.

따라서 이후 모델 학습에서는 클래스 가중치 조정, 적절한 평가 지표 선택, 그리고 Random Forest 등 전통적인 머신러닝 모델과의 비교가 필요하다. 특히 `pha` 분류에서는 Accuracy보다 Recall, Precision, F1-score, PR-AUC를 함께 확인해야 한다.
